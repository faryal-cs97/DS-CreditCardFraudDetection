## FOUND OUTLIER AS DATA SHOWS REIGHT-SKEWED DISTRIBUTION
1. Detecting Outliers
a. Using Statistical Methods (IQR)
The Interquartile Range (IQR) is a robust method to detect outliers:

IQR: The range between the 25th percentile (Q1) and 75th percentile (Q3).
Any value outside the range [Q1 - 1.5 * IQR, Q3 + 1.5 * IQR] is considered an outlier.

```python
# Calculate Q1, Q3, and IQR
Q1 = data['Amount'].quantile(0.25)
Q3 = data['Amount'].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Detect outliers
outliers = data[(data['Amount'] < lower_bound) | (data['Amount'] > upper_bound)]
print(f"Number of outliers in 'Amount': {len(outliers)}")
```

b. Using Visualization
Boxplot: Highlights outliers visually.

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data['Amount'], color='lightblue')
plt.title("Boxplot of Transaction Amounts")
plt.show()
```

Histogram with Log Scale: Highlights extreme values.

```python
plt.figure(figsize=(8, 5))
plt.hist(data['Amount'], bins=50, color='blue', alpha=0.7, log=True)
plt.title("Histogram of Transaction Amounts (Log Scale)")
plt.xlabel("Transaction Amount")
plt.ylabel("Frequency (Log Scale)")
plt.show()
```

2. Handling Outliers
a. Removing Outliers
If the outliers are likely due to data entry errors or are irrelevant to the analysis, remove them:

```python
# Remove outliers
filtered_data = data[(data['Amount'] >= lower_bound) & (data['Amount'] <= upper_bound)]
print(f"Data after outlier removal: {filtered_data.shape}")
```

b. Transforming Outliers
When outliers are genuine (e.g., legitimate high-value transactions), transform them to reduce their influence:

Log Transformation:

```python
data['Amount_Log'] = np.log1p(data['Amount'])  # log(1 + x) to handle zeros
```

Capping: Set outliers to the nearest threshold

```python
data['Amount_Capped'] = np.where(data['Amount'] > upper_bound, upper_bound, 
                                 np.where(data['Amount'] < lower_bound, lower_bound, data['Amount']))
```

c. Using Robust Scaler for Preprocessing
To make the dataset robust against outliers, use scaling methods like RobustScaler:

```python
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()
data['Amount_Scaled'] = scaler.fit_transform(data[['Amount']])
```
